```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 50x50 image with all zeros
image_size = 50
image_array = np.zeros((image_size, image_size), dtype=np.uint8)

# Draw a diagonal line from the top-left to the bottom-right
for i in range(image_size):
    image_array[i, i] = 255  # Set the pixel value to 255 (white) for the line

# Plot the image
plt.imshow(image_array, cmap='gray', vmin=0, vmax=255)
plt.title('Image with Diagonal Line')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_line(image, start, end, color=255):
    """Draw a line on the image from `start` to `end` with a specific color."""
    rr, cc = line(start[0], start[1], end[0], end[1])
    image[rr, cc] = color

def plot_image_with_lines():
    image_size = 50
    image_array = np.zeros((image_size, image_size), dtype=np.uint8)
    
    # Draw a diagonal line
    draw_line(image_array, (0, 0), (image_size-1, image_size-1), color=255)
    
    # Draw a vertical line
    draw_line(image_array, (0, int(image_size/4)), (image_size-1, int(image_size/4)), color=128)
    
    # Draw a horizontal line
    draw_line(image_array, (int(image_size/2), 0), (int(image_size/2), image_size-1), color=200)
    
    # Draw a line with a different angle
    draw_line(image_array, (0, int(image_size/2)), (image_size-1, int(image_size/4)), color=64)
    
    plt.imshow(image_array, cmap='gray', vmin=0, vmax=255)
    plt.title('Image with Different Types of Lines')
    plt.axis('off')
    plt.show()

# Utility function to create line coordinates
from skimage.draw import line

# Call the function to plot the image
plot_image_with_lines()
```
